## 將量刑結果從判決書取出來
---
經過階段性的觀察，透過正規表達法一層一層的將真正所需的刑度（量刑）所抓出，過程分為四段：

1. 如果是上訴駁回就直接刪除


2. 從判決書所有的內文中先取出正文部分
    -  主文～事實及理由
    -  主文～犯罪事實
    -  主文～事實
    
    
3. 正文中若有「無罪」，代表該判決最終結果為0月


       
4. 從正文中取出真正判刑的段落
    -  處有期徒刑x年x月
    -  應執行有期徒刑x年x月
     
    
5. 從最後的段落中把時間資料（x年x月）留下
    - 不要的字詞、符號包含：有期徒刑應執行處，；。

### 初始資料：原本判決書的內文

```python
clean_data["content"][3]
```

### STEP 1 ：如果是上訴駁回就直接刪除

```python
record = []
for i in range(len(clean_data)):
    if "主文上訴駁回。" in clean_data["content"][i]:
        record.append(i)
        clean_data.drop(index = i, inplace = True)
clean_data = clean_data.reset_index(drop=True)
```

### STEP 2 ：擷取主文
---
    經觀察找到的規律為
    1. 主文～事實及理由
    2. 主文～犯罪事實
    3. 主文～事實

```python
case=re.findall("主文[^~]*事實及理由|主文[^~]*犯罪事實+|主文[^~]*事實",clean_data["content"][3])[0]
case
```

### STEP 3：若主文中包含「無罪」，代表該判決最終結果為0年（0月）

```python
if "無罪" in case:
        sentence_list.append("零月")
```

### STEP 4：擷取包含「有期徒刑」段落
---
    經過觀察找到的規律為
    1. 處有期徒刑x年x月
    2. 處有期徒刑x年
    3. 處有期徒刑x月
    4. 應執行有期徒刑x年x月
    5. 應執行有期徒刑x年
    6. 應執行有期徒刑x月

```python
sentence=re.findall("處有期徒刑[^年]年[^月]月[；|，|。]|處有期徒刑[^年]年[；|，|。]|處有期徒刑[^月]月[；|，|。]|應執行有期徒刑[^年]年[^月]月[；|，|。]|應執行有期徒刑[^年]年[；|，|。]|應執行有期徒刑[^月]月[；|，|。]",case)[0]
sentence
```

### STEP 5：擷取刑期（真正的時間資料）
---
- 不要的字詞、符號包含：有期徒刑應執行處，；。

```python
num_sentence=re.sub("[有期徒刑應執行處，；。]",'', sentence)
num_sentence
```

### 上述步驟整合：提取刑度

```python
sentence_list=[]
record = []

for i in range(len(clean_data)):
    case=re.findall("主文[^~]*事實及理由|主文[^~]*犯罪事實+|主文[^~]*事實",clean_data["content"][i])[0]
    if "無罪" in case:
        sentence_list.append("零月")
        
    elif "有期徒刑" not in case:
        record.append(i)
        
    else:
        sentence=re.findall("處有期徒刑[^年]年[^月]月[；|，|。]|處有期徒刑[^年]年[；|，|。]|處有期徒刑[^月]月[；|，|。]|應執行有期徒刑[^年]年[^月]月[；|，|。]|應執行有期徒刑[^年]年[；|，|。]|應執行有期徒刑[^月]月[；|，|。]",case)[0]
        num_sentence=re.sub("[有期徒刑應執行處，；。]",'', sentence)
        sentence_list.append(num_sentence)
for i in record:
    clean_data.drop(index = i, inplace = True)
    clean_data = clean_data.reset_index(drop=True)
        
clean_data["量刑結果"] = sentence_list
```

### 把量刑結果轉換為以月份為單位：result(month)
處理目標：壹年貳月→1年2月→14月

```python
chi_to_num = {
    "零":0,
    "壹":1,
    "貳":2,
    "參":3,
    "肆":4,
    "伍":5,
    "陸":6,
    "柒":7,
    "捌":8,
    "玖":9,
    "拾":10
}

def total_month(num_result):    
    right_word = []
    total_word = []
    for each in num_result:
        temp = ""

        if each[-2] == "+":
            temp = each[:-2]
            right_word.append(temp)
        else:
            temp = each[:-1]
            right_word.append(temp)


    for each in right_word:
        temp = ""
        for each_word in each:
            if each_word == "年":
                temp += "*12+"
            else:
                temp+=each_word
        total_word.append(eval(temp))
    return total_word

def to_num_df(df):
    num_result = []
    for each in range(len(df)):
        clean_result = ""
        temp = df["量刑結果"][each]


        if "拾" in temp:
            for i in temp:
                if i == "拾":
                    clean_result += str(chi_to_num[i])
                    clean_result += "+" 
                elif i == "年" or i == "月":
                    clean_result += i
                else:
                    clean_result += str(chi_to_num[i])
            num_result.append(clean_result)
        else:

            for i in temp:
                if type(i) == int:
                    clean_result += str(i)
                elif i in chi_to_num.keys():
                    clean_result += str(chi_to_num[i])
                elif i == "年" or i == "月":
                    clean_result += i
                else:
                    clean_result += str(i)
            num_result.append(clean_result)
    return total_month(num_result)

def final_data():
    month_list = to_num_df(clean_data)
    clean_data['result(months)'] = month_list
    clean_data.drop(columns = ["量刑結果","content"], inplace = True)
    return clean_data
```

## 計算整體罪行等級
---

### 先將四位律師的表格合併在一起

```python
L1 = pd.read_csv("林更祐-詐欺(最終data).csv", index_col=0)
L2 = pd.read_csv("林福容-詐欺(最終data).csv", index_col=0)
L3 = pd.read_csv("江宜蔚-詐欺(最終data).csv", index_col=0)
L4 = pd.read_csv("王品懿-詐欺(最終data).csv", index_col=0)
```

```python
alldata = pd.concat([L1,L2,L3,L4],axis=0).reset_index(drop=True)
alldata
```

### 取出四個等級的分界點

```python
alldata.describe()
```

### 得出來的結果為：
1. 第一等級：0個月～2個月
2. 第二等級：3個月～6個月
3. 第三等級：7個月～10個月
4. 第四等級：11個月以上

```python
PR_df = alldata.describe()
percentile25 = PR_df.loc["25%","result(months)"]
percentile50 = PR_df.loc["50%","result(months)"]
percentile75 = PR_df.loc["75%","result(months)"]
print(percentile25)
print(percentile50)
print(percentile75)
```

### 計算總體平均刑度

```python
def get_level_grade(data):
    level_grade = []
    grade1 = data["result(months)"]<=percentile25
    grade1_result = data[grade1]["result(months)"]
    if len(grade1_result) == 0:
        level_grade.append(0)
    else:
        gg1 = sum(grade1_result)/len(grade1_result)
        level_grade.append(gg1)

    cut1 = data["result(months)"]>percentile25
    cut2 = data["result(months)"]<=percentile50
    grade2 = data[(cut1&cut2)]
    grade2_result = grade2["result(months)"]
    if len(grade2_result) == 0:
        level_grade.append(0)
    else:
        gg2 = sum(grade2_result)/len(grade2_result)
        level_grade.append(gg2)

    cut3 = data["result(months)"]>percentile50
    cut4 = data["result(months)"]<=percentile75
    grade3 = data[(cut3&cut4)]
    grade3_result = grade3["result(months)"]
    if len(grade3_result) == 0:
        level_grade.append(0)
    else:
        gg3 = sum(grade3_result)/len(grade3_result)
        level_grade.append(gg3)

    grade4 = data["result(months)"]>percentile75
    grade4_result = data[grade4]["result(months)"]
    if len(grade4_result) == 0:
        level_grade.append(0)
    else:
        gg4 = sum(grade4_result)/len(grade4_result)
        level_grade.append(gg4)
    
    return level_grade
```

### 總體平均刑度結果

```python
all_level_grade = get_level_grade(alldata)
```

```python
all_level_grade_df = pd.DataFrame(columns = ["總體平均刑度"], index = ["等級1","等級2","等級3","等級4"])
all_level_grade_df["總體平均刑度"] = all_level_grade
all_level_grade_df
```

```python
#林更祐
L1_level_grade = get_level_grade(L1)
print("林更祐律師平均量刑",L1_level_grade)

#林福容
L2_level_grade = get_level_grade(L2)
print("林福容律師平均量刑",L2_level_grade)

#江宜蔚
L3_level_grade = get_level_grade(L3)
print("江宜蔚律師平均量刑",L3_level_grade)

#王品懿
L4_level_grade = get_level_grade(L4)
print("王品懿律師平均量刑",L4_level_grade)
```

```python
all_level = pd.DataFrame(columns = ["總體","林更祐","林福容","江宜蔚","王品懿"], index = ["等級1","等級2","等級3","等級4"])
all_level["總體"] = all_level_grade
all_level["林更祐"] = L1_level_grade
all_level["林福容"] = L2_level_grade
all_level["江宜蔚"] = L3_level_grade
all_level["王品懿"] = L4_level_grade
all_level
```

## 各律師的減刑表現
---

### 算出此律師在詐欺案件的整體表現並新增至DataFrame中

```python
def perform(lawyer_data,lawyer_name):
    lawyer_data = all_level[["總體",lawyer_name]]
    lawyer_data["表現"] = (lawyer_data[lawyer_name]-lawyer_data["總體"])/lawyer_data["總體"]
    lawyer_data["整體表現"] = ""
    lawyer_data.loc["等級1","整體表現"] = sum(lawyer_data["表現"])
    return lawyer_data
```

```python
wong_perform = perform(wong_data,"王品懿")
wong_perform
```
